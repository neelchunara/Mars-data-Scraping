````python
#get all the link on the page
anchors = soup.find_all('a')
all_links=set()

for link in anchors:
    if(link.get('href') != '#'):
        linkText = "https://static.bc-edx.com/data/web/mars_news/index.html" + link.get('href')
        all_links.add(link)
        print(linkText)
````

````python
# Select one news article with select_one()
slide_elem = soup.select_one('div.list_text')
````

````python
# Scrape the title of the news article
title_elem = slide_elem.find('div', class_='content_title')
print(title_elem)
````

````python
# Extract the text of the title with get_text()
title = title_elem.get_text()
print(title)
````

````python
# Scrape the preview of the news article
preview_elem = slide_elem.find('div', class_='article_teaser_body')
print(preview_elem)
````

````python
# Extract the text of the preview with get_text()
preview = preview_elem.get_text()
print(preview)
````
